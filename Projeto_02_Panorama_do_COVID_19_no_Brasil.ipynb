# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Os avanços na compreensão do COVID-19 têm sido notáveis, mas ainda persistem lacunas no conhecimento. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://www2.ufjf.br/progepe/wp-content/uploads/sites/28/2020/03/Covid-19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados foram retirados de um repositório do Github, disponíveis em [https://github.com/owid/covid-19-data/tree/master/public/data](https://github.com/owid/covid-19-data/tree/master/public/data). Os dados encontrados são provenientes do Our World in Data (OWID), uma publicação digital que busca obter progresso contra os maiores problemas enfrentados na qualidade de vida ao redor do mundo.

Por se tratar de uma doença nova, muitas perguntas ainda tem poucos dados para análise. Tratamentos e sequelas ainda não possuem um conhecimento mais aprofundado, que possa levar a soluções eficazes.

## Dicionário de Variáveis
- *iso_code* = Código de identificação de três dígitos para um páis

- *continent* = Continente

- *location* = Localização

- *date* = Data

- *total_cases* = Total de casos confirmados de Covid-19

- *new_cases* = Novos casos confirmados de Covid-19

- *total_deaths* = Total de mortes pelo vírus

- *new_deaths* = Novas mortes registradas

- *total_cases_per_million* = Total de casos por milhão de habitantes

- *new_cases_per_million* = Novos casos por milhão de habitantes

- *total_deaths_per_million* = Total de mortes por milhão de habitantes

- *new_deaths_per_million* = Novas mortes por milhão de habitantes

- *total_tests* = Total de testes para a Covid-19

- *new_tests* = Novos testes

- *total_tests_per_thousand* = Total de testes por mil habitantes

- *new_tests_per_thousand* = Novos testes por mil habitantes

- *new_tests_smoothed* = Novos testes relatados

- *new_tests_smoothed_per_thousand* = Novos testes relatados por mil habitantes

- *tests_units* = Unidades usadas pelo local para relatar seus dados de teste

- *stringency_index* =  Índice de Rigidez de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibição de viagens, reescalonado para um valor de 0 a 100 (100 = resposta mais rigorosa)

- *population* = População

- *population_density* = Densidade da população

- *median_age* = Idade média das pessoas contaminadas

- *aged_65_older* = Maiores de 65 anos de idade

- *aged_70_older* = Maiores de 70 anos de idade

- *gdp_per_capita* = Produto interno bruto em paridade de poder de compra (dólares internacionais constantes de 2011), ano mais recente disponível

- *extreme_poverty* = Pessoas em extrema pobreza

- *diabetes_prevalence* = Pessoas com prevalência de diabetes

- *female_smokers* = Mulheres fumantes

- *male_smokers* = Homens fumantes

- *handwashing_facilities* = Percentagem da população com instalações básicas de lavagem das mãos nas instalações, ano mais recente disponível

- *hospital_beds_per_thousand* = Número de camas de hospital por mil habitantes

## Evolução da Covid-19 no Mundo

```
# Plotar um gráfico de linhas para ver a evolução do total de casos de morte por COVID-19 no mundo.
evolucao_total_de_mortes = df.loc[df.location == "World", "total_deaths"].reset_index(drop=True)

fig,ax = plt.subplots(figsize=(10,10))
evolucao_total_de_mortes.plot(x = df.date, y = 'total de mortes')
ax.set_title("Evolução do total de mortes no mundo")
ax.set_xlabel("Dias")
ax.set_ylabel("Total de Mortes")
```

Serão analisadas as seguintes questões:

* Quando foi registrada a primeira morte no Brasil?
* Quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado?
* Análise de gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.

```
#Criar uma cópia do DataFrame apenas com o Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()
```

Após a criação do DataFrame apenas com os dados do Brasil, é possivel coletar informações de maneira mais facilitada, mas antes de iniciarmos a análise, vamos conhecer melhor nosso novo DataFrame.

```
df_brasil.head()
```

```
df_brasil.date.max()
```

Acima, verificamos que nosso dataset inclui datas desde 31-12-2019 até 15-06-2020.

Vamos verificar quando foram identificados o primeiro caso e a primeira morte no Brasil.

```
# Identificar quando foi registrada o primeiro caso de Covid-19 no Brasil
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
print('O primeiro caso foi registrado em {}'.format(df_brasil.date.loc[3154]))
```

```
# primeira morte
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
```

```
# Identificar quando foi registrada a primeira morte no Brasil
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)

print('A primeira morte foi registrada em {}'.format(df_brasil.date.loc[3175]))
```

Um dado interessante de ser visualizado é o tempo entre a identificação do primeiro caso e o registro da primeira morte.

```
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# calculando tempo desde o primeiro caso até a primeira morte
from datetime import date
delta = df_brasil.date.loc[3175] - df_brasil.date.loc[3154]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(delta.days))
```

Para melhorar a visualização dos dados no Brasil, vamos plotar alguns gráficos, para melhor comparação entre o número de casos confirmados e o total de mortes registradas.

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(168))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica)')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```

## Conclusão

###Primeiro Caso e Primeira Morte no Brasil:

- O primeiro caso de COVID-19 no Brasil foi registrado em 26 de fevereiro de 2020, e a primeira morte ocorreu em 18 de março de 2020. Isso significa que passaram 21 dias desde o primeiro caso até a primeira morte no país.


###Evolução dos Casos e Mortes no Brasil:


- A evolução dos casos e mortes no Brasil segue uma tendência crescente ao longo do tempo. Observamos que, em escala linear, ambos os gráficos apresentam um aumento contínuo. Ao utilizar uma escala logarítmica, é possível visualizar de forma mais clara o crescimento exponencial inicial dos casos e mortes, indicando a importância das medidas de contenção.


###Comparação Internacional:

- Os Estados Unidos e o Brasil se destacam em termos de total de casos e mortes, o que pode ser atribuído à alta densidade populacional e extensão territorial desses países. É importante ressaltar que o número de casos por milhão de habitantes coloca o Brasil em uma posição menos crítica em comparação com países menores como San Marino e o Vaticano.


###Impacto nas Micronações:

- San Marino e o Vaticano, apesar de serem micronações com populações muito pequenas, apresentam altos números de casos por milhão de habitantes. Isso pode estar relacionado à proximidade geográfica com países fortemente afetados e à densidade populacional nessas áreas.


###Necessidade de Medidas Preventivas:

- A rápida disseminação do vírus, como observado nos primeiros casos no Brasil, destaca a importância de medidas preventivas, como distanciamento social, uso de máscaras e higienização frequente das mãos, para conter a propagação da doença.


###Limitações da Análise:

- É importante lembrar que esta análise é baseada em dados disponíveis até junho de 2020. A situação da COVID-19 está em constante evolução, e novos dados e descobertas podem ter surgido após esse período.


Essas conclusões fornecem uma visão geral da situação da COVID-19 no Brasil e no mundo até junho de 2020. É fundamental continuar acompanhando os dados e seguir as orientações das autoridades de saúde para enfrentar essa pandemia de forma eficaz.
